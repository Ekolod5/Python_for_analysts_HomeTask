### 1.1. Скачать следующие данные: kc-house-data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('./kc_house_data.csv')
df
```

### 1.2. Изучите стоимости недвижимости.

```python
plt.figure(figsize=(15, 10))
sns.histplot(df['price'])
plt.ticklabel_format(style='plain')
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Кол-во');
```

### 1.3. Изучите распределение квадратуры жилой.

```python
plt.figure(figsize=(6, 8))
sns.histplot(df['sqft_living'], bins=40)
plt.title('Распределение квадратуры жилой')
plt.xlabel('Кв. метры жилые')
plt.ylabel('Количество');
```

### 1.4. Изучите распределение года постройки.

```python
plt.figure(figsize=(8, 8))
sns.histplot(df['yr_built'], bins=40)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество');
```

## Задание 2.

### 2.1. Изучите распределение домов от наличия вида на набережную. Постройте график. Сделайте выводы.

```python
data = df['waterfront'].value_counts()
data
```

```python
plt.figure(figsize=(8, 8))
explode = [0, 0.7]
plt.pie(data, autopct='%1.1f%%', explode=explode, colors = ['#40E0D0', '#FFFF00'])
plt.title('Доля наличия вида на набережную')
plt.legend(['no', 'yes']);
```

Только у 0,8% домов наличие вида на набережную.

### 2.2. Изучите распределение этажей домов.

```python
data2 = df['floors'].value_counts()
data2
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x = data2.index, y = data2)
plt.title('Распределение этажей домов')
plt.xlabel('Кол-во этажей')
plt.ylabel('Кол-во')
plt.xticks(rotation=30);
```

```python
plt.figure(figsize=(10, 10))
expl = [0, 0, 0.04, 0.1, 0.3, 0.5]
plt.pie(data2.values, autopct='%1.1f%%', labels=data2.index, explode=expl)
plt.title('Доля этажности домов')
plt.legend(data2.index, loc='best');
```

В основном у домов 1(49,4%), 2(38,1%) этажа.

### 2.3. Изучите распределение состояния домов.

```python
data3 = df['condition'].value_counts()
data3
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x = data3.index, y = data3)
plt.title('Распределение состояния домов')
plt.xlabel('Состояние дома')
plt.ylabel('Кол-во');
```

```python
plt.figure(figsize=(8, 8))
exp2 = [0, 0, 0.1, 0.3, 0.6]
plt.pie(data3, autopct='%1.1f%%', labels=data3.index, explode=exp2, colors = ['orange', '#7FFFD4',  '#ADFF2F', '#FF00BB', 'Yellow'])
plt.title('Распределение состояния домов')
plt.legend(data3.index, loc='best');
```

В основном дома с состоянием - 3 (64,9%), 4(26,3%).

## Задание 3.

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='crest');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Чем больше жилплощадь дома, тем выше его стоимость.

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5)
plt.title('Distribution of price by grade')
plt.xlabel('price')
plt.ylabel('grade');
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5)
plt.title('Distribution of price by floors')
plt.xlabel('price')
plt.ylabel('floors');
```

Наибольшую стоимость имеют дома с количеством этажей 2,5.

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5)
plt.title('Distribution of price by view')
plt.xlabel('price')
plt.ylabel('view');
```

Наибольшую стоимость имеют дома с view - 4.

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), whis=1.5)
plt.title('Distribution of price by waterfront')
plt.xlabel('price')
plt.ylabel('waterfront');
```

```python
plt.figure(figsize=(10, 8))
sns.violinplot(x=df['price'], y=df['waterfront'].astype('str'), whis=1.5)
plt.title('Distribution of price by waterfront')
plt.xlabel('price')
plt.ylabel('waterfront');
```

Наибольшую стоимость имеют дома с наличием вида на набережную.
